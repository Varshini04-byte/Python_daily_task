```python
#menu-driven
while True:
    print("\n1.Add")
    print("2.subtract")
    print("3.Multiply")
    print("4.Divide")
    print("5.Exit")
    choice=int(input("Enter your choice: "))
    if choice==1:
        a=int(input("Enter first number: "))
        b=int(input("Enter second number: "))
        print("Sum ",a+b)
        break
    elif choice==2:
        a=int(input("Enter first number: "))
        b=int(input("Enter second number: "))
        print("Difference ",a-b)
        break
    elif choice==3:
        a=int(input("Enter first number: "))
        b=int(input("Enter second number: "))
        print("Multiply ",a*b)
        break
    elif choice==4:
        a=int(input("Enter first number: "))
        b=int(input("Enter second number: "))
        print("Divide ",a/b)
        break
    elif choice==5:
        print("Program exit")
        break
    else:
        print("Invalid choice")
        
```

```python
#ATM System
balance=2000
while True:
    print("\n1.check balance")
    print("2.Deposit money")
    print("3.withdraw money")
    print("4.Exit")
    choice=int(input("enter your choice: "))
    if choice==1:
        print("check balance ",balance)
    elif choice==2:
        amt=int(input("Enter your amount: "))
        balance=balance+amt
        print("update balance ",balance)
    elif choice==3:
        amt=int(input("Enter your amount: "))
        if amt<=balance:
            balance=balance-amt
            print("update balance ",balance)
        else:
            print("Insufficient balance")
    elif choice==4:
        print("Exit")
        break
    else:
        print("Invalid choice")
        
        
```
