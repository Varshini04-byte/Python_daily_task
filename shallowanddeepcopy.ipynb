```python
#shallow copy
import copy
original=[1,2,[3,4],5]
shallow=copy.copy(original)
shallow[2][0]=99
print("original object:",original)
print("shallow object:",shallow)
```

```python
#deep copy
import copy
original=[1,2,[3,4],5]
deep=copy.deepcopy(original)
deep[2][0]=99
print("original object:",original)
print("deep object:",deep)
```

```python
#nested shallow copy
import copy
original=[[1,2],[3,4]]
shallow=copy.copy(original)
shallow[1][0]=99
print("original object:",original)
print("shallow object:",shallow)
```

```python
#nested deep copy
import copy
original=[[1,2],[3,4],[5,6]]
deep=copy.deepcopy(original)
deep[1][1]=9
print("original object:",original)
print("deep object:",deep)
```

```python
# task 7:

def manual_create_deep(lst):

    if isinstance(lst,list):

        return [manual_create_deep(item) for item in lst]

    else :
        return lst


# original list :
original_list = [1,2,[3,4],[5,[6,7,8]]]

# calling the function
deep_copy_list=manual_create_deep(original_list)


# modify the the deep copy list 
deep_copy_list[3][1][0] = 50

print("Original List",original_list)
print("Copy List",deep_copy_list)
```

```python
# 9 :

def detect_copy(original):

    # create the shallow copy
    shallow_copy = copy.copy(original)

    # create the deep copy 
    deep_copy =  copy.deepcopy(original)
    # modify the shallow copy 
    shallow_copy[2][1] = 100
    #modify the the deep copy
    deep_copy[1][0] = 200

    if shallow_copy == original:

        return f"Shallow Copy : {shallow_copy}"
        
    if deep_copy != original:
        return  f"Deep Copy : {deep_copy}"
# original list 
original = [[1,2],[3,4],[5,6]]
# call the function 
detect_copy(original)
```

```python
#Level:5
#Task23:
class BusReservation:
    def __init__(self, bus_name, total_seats):
        self.bus_name = bus_name
        self.total_seats = total_seats
        self.booked_seats = 0

    def reserve(self, number):
        if self.booked_seats + number <= self.total_seats:
            self.booked_seats += number
            print("Reservation Successful")
            print("Remaining Seats:", self.total_seats - self.booked_seats)
        else:
            print("Reservation Failed")
            print("Not enough seats available")

b = BusReservation("Express Bus", 50)

b.reserve(20)
b.reserve(15)
b.reserve(20)
```

```python
#Level:5
#Task25:
class ShoppingCart:
    def __init__(self, customer_name, wallet_balance):
        self.customer_name = customer_name
        self.cart_total = 0
        self.wallet_balance = wallet_balance

    def add_item(self, price):
        self.cart_total += price
        print("Item Added")
        print("Cart Total:", self.cart_total)

    def remove_item(self, price):
        if self.cart_total - price < 0:
            print("Cart total cannot become negative")
        else:
            self.cart_total -= price
            print("Item Removed")
            print("Cart Total:", self.cart_total)

    def checkout(self):
        if self.cart_total == 0:
            print("Cart is empty")
        elif self.wallet_balance >= self.cart_total:
            self.wallet_balance -= self.cart_total
            print("Payment Successful")
            print("Remaining Wallet Balance:", self.wallet_balance)
            self.cart_total = 0
        else:
            print("Insufficient Wallet Balance")

c = ShoppingCart("Sushma", 5000)

c.add_item(1000)
c.add_item(1500)

c.remove_item(500)

c.checkout()

c.checkout()
```
