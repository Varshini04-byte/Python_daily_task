```python
def calculate(a,b, operation):
    if operation == "add":
        return a+b
    elif operation == "sub":
        return a-b
    elif operation == "mul":
        return a*b
    elif operation == "div":
        return a/b
calculate(a=10,b=5, operation="sub")
```

```python
#args(Multiple Positional Arguments)
def add(*numbers):
    c=sum(numbers)
    return c
add(20,30,50,60)
```

```python
def print_all(*args):
    return args
print_all(10.2,10,"hello")
```

```python
def find_max(*args):
    return max(args)
find_max(20,34,57,2,5)
```

```python
def concatenate(*args):
    return " ".join(args)
concatenate("Hello","good","morning","to","all")
```

```python
def multiply(*args):
    return multiply(args)
multiply(2,3,5,6)
```
