```python
#Hardlevel:
#Task1:
class Employee:
    total_employees = 0
    last_updated_salary = 0

    def __init__(self, salary):
        if salary < 5000:
            self.__salary = 5000
        else:
            self.__salary = salary

        Employee.total_employees += 1
        Employee.last_updated_salary = self.__salary

    # Getter
    def get_salary(self):
        return self.__salary

    # Setter
    def set_salary(self, new_salary):
        if new_salary < 5000:
            print("Salary cannot be less than 5000")
        elif new_salary > self.__salary * 1.20:
            print("Salary cannot be increased by more than 20% at once")
        else:
            self.__salary = new_salary
            Employee.last_updated_salary = new_salary
            print("Salary updated successfully")

    @classmethod
    def get_total_employees(cls):
        return cls.total_employees


# Example
e1 = Employee(10000)
e2 = Employee(8000)

print("Salary:", e1.get_salary())

e1.set_salary(11000)   # Valid
print("Updated Salary:", e1.get_salary())

e1.set_salary(15000)   # More than 20% increase

e1.set_salary(4000)    # Less than 5000

print("Last Updated Salary:", Employee.last_updated_salary)
print("Total Employees:", Employee.get_total_employees())
```

```python
#Hardlevel:
#Task2:
class BankAccount:
    total_accounts = 0
    history = []

    def __init__(self, balance):
        if balance < 1000:
            self.__balance = 1000
        else:
            self.__balance = balance
            if self.__balance > 100000:
                self.__balance += self.__balance * 0.05

        BankAccount.total_accounts += 1
        BankAccount.history.append(self.__balance)

    # Getter
    def get_balance(self):
        return self.__balance

    # Setter
    def set_balance(self, new_balance):
        if new_balance < 1000:
            print("Balance cannot go below minimum balance (1000)")
        else:
            if new_balance > 100000:
                new_balance += new_balance * 0.05
                print("5% bonus added")

            self.__balance = new_balance
            BankAccount.history.append(self.__balance)
            print("Balance updated successfully")

    @classmethod
    def show_total_accounts(cls):
        return cls.total_accounts

    @classmethod
    def show_history(cls):
        return cls.history


# Example
b1 = BankAccount(50000)
b2 = BankAccount(120000)

print("Balance:", b1.get_balance())

b1.set_balance(150000)
print("Updated Balance:", b1.get_balance())

b1.set_balance(500)

print("History:", BankAccount.show_history())
print("Total Accounts:", BankAccount.show_total_accounts())
```
