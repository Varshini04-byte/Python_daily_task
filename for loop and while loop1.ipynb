```python
#menu-driven
while True:
    print("\n1.Add")
    print("2.subtract")
    print("3.Exit")
    choice=int(input("Enter your choice: "))
    if choice==1:
        a=int(input("Enter first number: "))
        b=int(input("Enter second number: "))
        print("Sum ",a+b)
        break
    elif choice==2:
        a=int(input("Enter first number: "))
        b=int(input("Enter second number: "))
        print("Difference ",a-b)
        break
    elif choice==3:
        print("Program exit")
        break
    else:
        print("Invalid choice")
        
```
