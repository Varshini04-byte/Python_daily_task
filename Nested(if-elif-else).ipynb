```python
a=int(input())
b=int(input())
c=int(input())
if a==b==c:
    print("Equal")
else:
    if a>b and a>c:
        print("a is largest",a)
        if a%2==0:
            print("Even number")
        else:
            print("Odd number")
    elif b>a and b>c:
        print("b is largest",b)
        if b%2==0:
            print("Even number")
        else:
            print("Odd number")
    else:
        print("c is largest",c)
        if c%2==0:
            print("Even number")
        else:
            print("Odd number")
```
