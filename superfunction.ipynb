```python

#Task1:
from abc import ABC, abstractmethod 
class Shape(ABC) :
        
    @abstractmethod
    def calculate_area(self):
        pass
    @abstractmethod
    def calculate_perimeter(self):
        pass
class Circle(Shape):
    def __init__(self,radius):
        self.radius=radius
    def calculate_area(self):
        return 3.14*self.radius*self.radius
           
    def calculate_perimeter(self):
        return 2*3.14*self.radius
class Rectangle(Shape):
    def __init__(self,length,breadth):
        self.length=length
        self.breadth=breadth
    
    def calculate_area(self):
        return self.length*self.breadth
            
    def calculate_perimeter(self):
        return 2*(self.length+self.breadth)
c=Circle(5)
print("Circle area:",c.calculate_area())
print("Circle perimeter:",c.calculate_perimeter())
r=Rectangle(10,5)
print("rectangle area:",r.calculate_area())
print("rectangle perimeter:",r.calculate_perimeter())
```

```python
#Task14:
from abc import ABC, abstractmethod
import math

# Abstract class
class Shape3D(ABC):

    @abstractmethod
    def volume(self):
        pass

    @abstractmethod
    def surface_area(self):
        pass
# Cube subclass
class Cube(Shape3D):

    def __init__(self, side):
        self.side = side

    def volume(self):
        return self.side ** 3
    def surface_area(self):
        return 6 * self.side ** 2
# Sphere subclass
class Sphere(Shape3D):

    def __init__(self, radius):
        self.radius = radius

    def volume(self):
        return (4 / 3) * math.pi * self.radius ** 3
    def surface_area(self):
        return 4 * math.pi * self.radius ** 2


# Objects
cube = Cube(4)
sphere = Sphere(3)

# Output
print("Cube Volume:", cube.volume())
print("Cube Surface Area:", cube.surface_area())

print("Sphere Volume:", sphere.volume())
print("Sphere Surface Area:", sphere.surface_area())

```

```python
#Task30:
from abc import ABC, abstractmethod

# Abstract class
class Report(ABC):
    def __init__(self,title):
        self.title=title
    def show_header(self):
        print("Report Title:",self.title)
    @abstractmethod
    def generate_report(self):
        pass
# subclass
class PDFReport(Report):
    def __init__(self,title):
        super().__init__(title)

    def generate_report(self):
        super().show_header()
        print(f"PDF report generated ")
class HTMLReport(Report):
    def __init__(self,title):
        super().__init__(title)

    def generate_report(self):
        super().show_header()
        print(f"PDF report generated ")




# Objects
p=PDFReport("Annual Report")
p.generate_report()
print()
h=HTMLReport("Sales Report")
h.generate_report()
```
